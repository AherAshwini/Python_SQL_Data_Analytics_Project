```python
#Extract the dataset using zipfile library.
import zipfile
zip_ref = zipfile.ZipFile('retail-orders.zip')
zip_ref.extractall()
zip_ref.close
```

```python
#Read dataset into Pandas dataframe.
import pandas as pd
df = pd.read_csv('orders.csv', na_values = ['unknown', 'Not Available']) #Replacing Not Available, unknown values with NaN.
df.columns = df.columns.str.replace(' ','_') #Formatting the column names for consistency.
df.columns = df.columns.str.lower()
```

```python
#Derive new columns as discount, sale price and profit.
df['discount'] = df['list_price']*df['discount_percent']*0.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
```

```python
#Convert order date from object data type to datetime.
df['order_date'] = pd.to_datetime(df['order_date'], format = '%Y-%m-%d')
```

```python
#Drop cost_price, list_price and discount_percent columns.
df = df.drop(['cost_price', 'list_price', 'discount_percent'], axis =1)
df
```
